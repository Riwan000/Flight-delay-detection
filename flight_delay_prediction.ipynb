```python
def create_shuffled_subset_and_save(df, num_rows=2000, output_folder='./'):
    """
    Creates a shuffled subset of the original DataFrame containing a specified number of rows and saves it as a CSV.

    Parameters:
    df (pd.DataFrame): The original DataFrame.
    num_rows (int): The number of rows for the subset (default is 2000).
    output_folder (str): The folder where the CSV will be saved (default is the current directory).

    Returns:
    pd.DataFrame: A shuffled subset of the original DataFrame with the specified number of rows.
    """
    # Check if the DataFrame has enough rows
    if len(df) < num_rows:
        raise ValueError(f"The DataFrame only has {len(df)} rows. Cannot create a subset of {num_rows} rows.")
    
    # Shuffle the DataFrame
    shuffled_df = df.sample(frac=1, random_state=42).reset_index(drop=True)  # Reset index after shuffling
    
    # Create the subset by selecting the first num_rows rows
    subset_df = shuffled_df.iloc[:num_rows].copy()
    
    # Create the output filename including the num_rows
    output_filename = f"airline_{num_rows}_rows.csv"
    
    # Save the subset DataFrame to a CSV file
    subset_df.to_csv(output_filename, index=False)  # Save without the index
        
    return subset_df
```

```python
# Function for creating a table of columns with missing values and their details.

def columns_with_missing_value(df_2000):
    '''
    Creates a dataframe with name of columns with missing data, number of missing values in each and the datatype of those columns.
    '''

    # List of columns with missing values
    missing_value_cols = [columns for columns in df_2000.columns if df_2000[columns].isnull().sum() > 0]

    # Create a DataFrame for the number of missing values
    number_of_missing = {'Columns': missing_value_cols,
                        'Number of missing values': [df_2000[columns].isnull().sum() for columns in missing_value_cols]}

    missing_list_df = pd.DataFrame(number_of_missing)

    # Create a DataFrame for the datatypes of columns with missing values
    missing_type_df = pd.DataFrame({
        'Columns': missing_value_cols,
        'Datatype': [df_2000[columns].dtypes for columns in missing_value_cols]
    })

    # Combine both DataFrames
    df_missing_details = pd.merge(missing_list_df, missing_type_df, on='Columns')

    # Return the result
    return df_missing_details
```

```python
def majority_missing(df_2000):
    # Identify features with more than 95% missing values
    threshold = 0.95 * len(df_2000)

    df_missing_details = columns_with_missing_value(df_2000)

    # Create a list of columns with a majority of missing values
    features_with_majority_missing = [
        column for column, missing_count in zip(df_missing_details['Columns'], df_missing_details['Number of missing values']) 
        if missing_count > threshold
    ]

    return features_with_majority_missing
```

```python
# Function to compute the midpoint of a time block
def get_midpoint(time_block):
    if pd.isna(time_block):
        return np.nan
    start, end = map(int, time_block.split('-'))
    midpoint = (start + end) // 2  # Calculate the midpoint
    return midpoint

# Fill null values in 'ArrTime' with the computed midpoint directly
df_2000['ArrTime'] = df_2000.apply(
    lambda row: get_midpoint(row['ArrTimeBlk']) if pd.isna(row['ArrTime']) else row['ArrTime'], axis=1
)
```

```python
def calculate_ground_time(row):
    # Use Div1TotalGTime if available
    if not np.isnan(row['Div1TotalGTime']):
        return row['Div1TotalGTime']
    # Otherwise, calculate from Div1WheelsOff and Div1WheelsOn
    elif not np.isnan(row['Div1WheelsOff']) and not np.isnan(row['Div1WheelsOn']):
        return row['Div1WheelsOff'] - row['Div1WheelsOn']
    # Return NaN if values are insufficient
    else:
        return np.nan

# Apply the function to calculate total ground time
df_2000['TotalAddGTime'] = df_2000.apply(calculate_ground_time, axis=1)
```

```python
# Define the function to calculate arrival delay for a single row
def calculate_arrival_delay(row):
    try:
        # Convert CRSArrTime and ArrTime to minutes past midnight
        CRSArrTime = (int(row['CRSArrTime']) // 100) * 60 + (int(row['CRSArrTime']) % 100)
        ArrTime = (int(row['ArrTime']) // 100) * 60 + (int(row['ArrTime']) % 100)

        # Adjust for overnight flights
        if ArrTime < CRSArrTime:
            ArrTime += 1440  # Add 24 hours in minutes

        # Calculate and return delay
        return ArrTime - CRSArrTime
    except ValueError:
        # Handle potential conversion errors
        return None

# Apply the function only to rows where 'ArrDelay' is null
df_2000.loc[df_2000['ArrDelay'].isna(), 'ArrDelay'] = df_2000[df_2000['ArrDelay'].isna()].apply(calculate_arrival_delay, axis=1)
```

```python
def fill_missing_arr_delay_minutes(df):
    def calculate_arrival_delay_minutes(row):
        CRSArrTime = (int(row['CRSArrTime']) // 100) * 60 + (int(row['CRSArrTime']) % 100)
        ArrTime = (int(row['ArrTime']) // 100) * 60 + (int(row['ArrTime']) % 100)

        if ArrTime < CRSArrTime:
            ArrTime += 1440

        arr_delay = ArrTime - CRSArrTime
        arr_delay_minutes = max(0, arr_delay)

        return arr_delay_minutes

    rows_with_nulls = df[df['ArrDelayMinutes'].isna()]

    for idx, row in rows_with_nulls.iterrows():
        arr_delay_minutes = calculate_arrival_delay_minutes(row)
        df.at[idx, 'ArrDelayMinutes'] = arr_delay_minutes

    return df

# Apply the function
df_2000 = fill_missing_arr_delay_minutes(df_2000)
```

```python
def fill_missing_arrival_delay_groups(df):
    def calculate_arrival_delay_group(row):
        CRSArrTime = (int(row['CRSArrTime']) // 100) * 60 + (int(row['CRSArrTime']) % 100)
        ArrTime = (int(row['ArrTime']) // 100) * 60 + (int(row['ArrTime']) % 100)

        if ArrTime < CRSArrTime:
            ArrTime += 1440

        arr_delay = ArrTime - CRSArrTime

        if arr_delay < -15:
            arrival_delay_group = -2
        elif -15 <= arr_delay <= -1:
            arrival_delay_group = -1
        elif 0 <= arr_delay <= 14:
            arrival_delay_group = 0
        else:
            arrival_delay_group = (arr_delay // 15) + 1

        return arrival_delay_group

    rows_with_nulls = df[df['ArrivalDelayGroups'].isna()]

    for idx, row in rows_with_nulls.iterrows():
        arrival_delay_group = calculate_arrival_delay_group(row)
        df.at[idx, 'ArrivalDelayGroups'] = arrival_delay_group

    return df

# Apply the function
df_2000 = fill_missing_arrival_delay_groups(df_2000)
```

```python
def calculate_elapsed(dep, arr):
    if pd.isna(dep) or pd.isna(arr):
        return None
    if arr < dep:  # Adjust for overnight flights
        return (arr + 2400 - dep) % 2400
    return arr - dep

df_2000['EstimatedElapsedTime'] = df_2000.apply(
    lambda row: calculate_elapsed(row['DepTime'], row['ArrTime']),
    axis=1
)

# Fill missing values in ActualElapsedTime
df_2000['ActualElapsedTime'].fillna(df_2000['EstimatedElapsedTime'], inplace=True)
```

```python
def calculate_airtime(wheels_off, wheels_on):
    if pd.isna(wheels_off) or pd.isna(wheels_on):
        return None
    if wheels_on < wheels_off:  # Adjust for overnight flights
        return (wheels_on + 2400 - wheels_off) % 2400
    return wheels_on - wheels_off

df_2000['CalculatedAirTime'] = df_2000.apply(
    lambda row: calculate_airtime(row['WheelsOff'], row['WheelsOn']),
    axis=1
)

# Fill missing values in AirTime
df_2000['AirTime'].fillna(df_2000['CalculatedAirTime'], inplace=True)
```

#### **Transforming categorical values into numerical values.**

```python
# Identify categorical and numerical columns
categorical_columns = df_2000.select_dtypes(include=['object', 'category']).columns
numerical_columns = df_2000.select_dtypes(include=['int64', 'float64']).columns
datetime_columns = df_2000.select_dtypes(include=['datetime64']).columns

# Convert datetime columns into numerical features
for col in datetime_columns:
    df_2000[f"{col}_hour"] = pd.to_datetime(df_2000[col]).dt.hour
    df_2000[f"{col}_dayofweek"] = pd.to_datetime(df_2000[col]).dt.dayofweek
    df_2000[f"{col}_month"] = pd.to_datetime(df_2000[col]).dt.month

# Drop original datetime columns
X = df_2000.drop(columns=datetime_columns)

# Define column transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numerical_columns),
        ('cat', OneHotEncoder(drop='first', sparse_output=False), categorical_columns)
    ],
    remainder='drop'  # Ensure no extra columns are retained
)

# Transform data
X_transformed = preprocessor.fit_transform(X)

# Check output shape
print(f"Shape of X_transformed: {X_transformed.shape}")

# Extract feature names
feature_names = (
    numerical_columns.tolist() +
    preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_columns).tolist()
)

# Verify feature names length matches transformed data
print(f"Number of feature names: {len(feature_names)}")

# Convert to DataFrame
transformed_df_2000 = pd.DataFrame(X_transformed, columns=feature_names)
```
